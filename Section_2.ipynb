```python
import pandas as pd
import numpy as np

# Step 1: Load the data
df = pd.read_csv("https://raw.githubusercontent.com/YashJ1825/Database-and-analitics/refs/heads/main/bookstore_transactions.csv")

# Step 2: View the first few rows
print(df.head())

```

```python
#1.	Check for Duplicate Records
print("Duplicate rows:", df.duplicated().sum())

```

```python
#2.	Check for Missing Values
print("Missing values per column:\n", df.isnull().sum())
```

```python

#3.	Check that Total Rows Match Expected Count
print("Dataset shape:", df.shape)
print(df.head())
```

```python
# Check structure
print("Shape of dataset:", df.shape)
print("Column names:", df.columns.tolist())
```

```python


# Check for uniqueness of transaction IDs
print("Unique transactions:", df['Transaction ID'].nunique())
print("Total records:", len(df))
```

```python
#Using Pandas:
print("Mean:\n", df.mean(numeric_only=True))
print("Median:\n", df.median(numeric_only=True))
print("Standard Deviation:\n", df.std(numeric_only=True))
print("Minimum:\n", df.min(numeric_only=True))
print("Maximum:\n", df.max(numeric_only=True))
print("Quantiles:\n", df.quantile([0.25, 0.5, 0.75], numeric_only=True))
```

```python
# Using NumPy:
print("Mean (NumPy):", np.mean(df["Discount Applied"]))
print("Median (NumPy):", np.median(df["Discount Applied"]))
print("Standard Deviation (NumPy):", np.std(df["Discount Applied"]))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Reload the dataset in case it's missing from memory
bookstore_df = pd.read_csv("https://raw.githubusercontent.com/YashJ1825/Database-and-analitics/refs/heads/main/bookstore_transactions.csv")

# Select numerical columns
numeric_cols = ['Stock Before', 'Stock After', 'Loyalty Points', 'Discount Applied']

# Create pairplot
sns.pairplot(bookstore_df[numeric_cols])
plt.suptitle("Pairplot of Numerical Features", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(bookstore_df[numeric_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.scatterplot(x='Discount Applied', y='Loyalty Points', data=bookstore_df)
plt.title("Loyalty Points vs Discount Applied")
plt.xlabel("Discount Applied")
plt.ylabel("Loyalty Points")
plt.grid(True)
plt.show()
```
